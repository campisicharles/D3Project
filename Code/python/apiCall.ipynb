```python
import pandas as pd
import requests
from config import api_key
```

```python
csv_path = "../../Documents/foodList.csv"

ndbid_df = pd.read_csv(csv_path, dtype=str)
ndbid_df
```

```python
ids = ndbid_df["ID"]
y=0
foodNutrientData = []
for x in ids:


    ndbno = ndbid_df["ID"][y]
    target_url = 'https://api.nal.usda.gov/ndb/V2/reports?ndbno={0}&type=b&format=json&api_key={1}'.format(ndbno,api_key)
    food_data = requests.get(target_url).json()
    print(ndbno)
    name = food_data['foods'][0]['food']['desc']['name']
    calories = food_data['foods'][0]['food']['nutrients'][0]['value']
    protein = food_data['foods'][0]['food']['nutrients'][1]['value']
    fat = food_data['foods'][0]['food']['nutrients'][2]['value']
    carb = food_data['foods'][0]['food']['nutrients'][3]['value']
    fiber = food_data['foods'][0]['food']['nutrients'][4]['value']
    sugar = food_data['foods'][0]['food']['nutrients'][5]['value']

    foodNutrientData.append({"Name": name, 
                   "Energy (kcal)": calories, 
                   "Protein (g)": protein, 
                   "Fat (g)": fat,
                   "Carbohydrates (g)": carb,
                   "Fiber (g)": fiber,
                   "Sugar (g)": sugar})
    y+=1
    
```

```python
food_df = pd.DataFrame(foodNutrientData)
food_df.head()
```

```python
food_df.to_csv("nutritionData.csv", index_label="Food_ID")
```
