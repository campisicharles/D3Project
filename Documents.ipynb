```python
envi_pd = pd.read_csv('../Data/environmental_impact_of_food_resultsNU.csv')
envi_pd.head()
```

```python
list(envi_pd)
```

```python
# Create new table
select_envi = envi_pd[["Unnamed: 0", "Nutr. Units / FU", "Unnamed: 5",
                           "Unnamed: 11", "Unnamed: 23", "Unnamed: 29", "Unnamed: 35"]]
select_envi.head()
```

```python
rename = select_envi.rename(columns={"Unnamed: 0": "Product", "Unnamed: 5": "Land_Use_m2/nutritional unit","Unnamed: 11":"GHG_Emissions_kg_CO2eq/NU",
                          "Unnamed: 23":"GHG_Emissions_kg_SO2eq/NU", "Unnamed: 29" :"GHG_Emissions_kg_PO4eq/NU","Unnamed: 35":"Freshwater_Withdrawals_L/NU"})
rename.head()
```

```python
clean_envi_data = rename.drop([rename.index[0] , rename.index[1]])
clean_envi_data.head()
```

```python
clean_envi = clean_envi_data.reset_index()
clean_envi.head()
```

```python
final_envi = clean_envi.dropna()

final_envi2 = final_envi.drop(['index'], axis=1)
final_envi2.head()
```

```python
final_envi2 = final_envi2.rename(columns={"Nutr. Units / FU": "NutrUnits", 
                                      "Land_Use_m2/nutritional unit": "LandUse",
                                      "GHG_Emissions_kg_CO2eq/NU": "CO2Emissions",
                                      "GHG_Emissions_kg_SO2eq/NU": "SO2Emissions",
                                      "GHG_Emissions_kg_PO4eq/NU": "PO4Emissions",
                                      "Freshwater_Withdrawals_L/NU": "FreshwaterWithdrawals"})

final_envi2.head()
```

```python
final_envi2.to_csv("/Data/clean_environment_data.csv", index=False)
```
